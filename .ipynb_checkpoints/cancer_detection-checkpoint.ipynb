```python
import pandas as pd
import sklearn 
import numpy as np


names=['std_id','clump_thick','cell_size','cell_shape','marginal_adhesion','sgl_epi_cell_size','bare_nulei','bland_chromatin','normal_nuclei','mitoses','class']
df=pd.read_csv("breast-cancer-wisconsin.data",names=names,na_values=['?'])

df.head()
```

```python
df.describe()
```

```python
df.loc[df['bare_nulei'].isnull()]
```

```python
df.dropna(inplace=True)
df.describe()
```

```python
features=df.columns[1:11].values
feature_df=np.array(df[features])
clas=np.array(df[['class']])

print(feature_df[0])
print(clas[0])
```

```python
from sklearn.model_selection import train_test_split
from sklearn import tree

x_train,x_test,y_train,y_test=train_test_split(feature_df,clas,test_size=0.2,random_state=1)

des_tree=tree.DecisionTreeClassifier(random_state=1)
des_tree.fit(x_train,y_train)
```

```python
des_tree.score(x_test,y_test)
```

```python
pred=des_tree.predict(x_test)
for i in range(len(pred)):
    print(pred[i]," ",x_test[i]," ",y_test[i][0])
```
